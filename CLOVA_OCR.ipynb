```python
import base64
import json
import os
import uuid
import time

#    Base64 
image_path = "_resume.jpg"
with open(image_path, "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read()).decode("utf-8")

# JSON Body 
json_body = {
    "version": "V2",
    "requestId": str(uuid.uuid4()),  #   
    "timestamp": int(time.time() * 1000),  #   
    "images": [
        {
            "format": "jpg",
            "name": "_resume.jpg",
            "data": encoded_string,
            "templateIds": [30424],  #  ID
            "templateArgs": {}
        }
    ]
}

# JSON Body  ()
print(json.dumps(json_body, indent=4))

# JSON   ()
save_dir = 'resume_data'
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, 'image_json_body.json')

with open(save_path, 'w', encoding='utf-8') as outfile:
    json.dump(json_body, outfile, ensure_ascii=False, indent=4)

print(f"JSON  '{save_path}'  .")
```

```python
import json
import os
from pprint import pprint

# JSON  
with open('.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 'fields'  
fields = data['images'][0]['fields']

# 'inferText' 'name' 
resume_data = [{'': field['name'], '': field['inferText']} for field in fields]

#   
pprint(resume_data)

#   
save_dir = 'resume_data'
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, '_data.json')

#   JSON  
with open(save_path, 'w', encoding='utf-8') as outfile:
    json.dump(resume_data, outfile, ensure_ascii=False, indent=4)

print(f"  '{save_path}'  .")
```
