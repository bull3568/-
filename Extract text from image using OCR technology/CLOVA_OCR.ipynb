```python
import base64
import json
import os
import uuid
import time

# 이미지 파일을 읽어 Base64로 인코딩
image_path = "이진아_resume.jpg"
with open(image_path, "rb") as image_file:
    encoded_string = base64.b64encode(image_file.read()).decode("utf-8")

# JSON Body 생성
json_body = {
    "version": "V2",
    "requestId": str(uuid.uuid4()),  # 고유한 식별자 생성
    "timestamp": int(time.time() * 1000),  # 현재 시간 밀리초
    "images": [
        {
            "format": "jpg",
            "name": "이진아_resume.jpg",
            "data": encoded_string,
            "templateIds": [30424],  # 템플릿 ID
            "templateArgs": {}
        }
    ]
}

# JSON Body 출력 (필요시)
print(json.dumps(json_body, indent=4))

# JSON 파일로 저장 (필요시)
save_dir = 'resume_data'
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, 'image_json_body.json')

with open(save_path, 'w', encoding='utf-8') as outfile:
    json.dump(json_body, outfile, ensure_ascii=False, indent=4)

print(f"JSON 데이터가 '{save_path}' 파일로 저장되었습니다.")
```

```python
import json
import os
from pprint import pprint

# JSON 파일 읽기
with open('이진아.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 'fields' 파트만 추출
fields = data['images'][0]['fields']

# 'inferText'와 'name' 추출
resume_data = [{'항목': field['name'], '내용': field['inferText']} for field in fields]

# 추출한 내용 출력
pprint(resume_data)

# 저장 경로 지정
save_dir = 'resume_data'
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, '이진아_data.json')

# 추출한 데이터를 JSON 파일로 저장
with open(save_path, 'w', encoding='utf-8') as outfile:
    json.dump(resume_data, outfile, ensure_ascii=False, indent=4)

print(f"추출한 데이터가 '{save_path}' 파일로 저장되었습니다.")
```
