```python

import json
import pandas as pd

# Load the JSON files
with open('./database/user_info.json', 'r', encoding='utf-8') as file:
    user_info = json.load(file)

with open('./database/company_info.json', 'r', encoding='utf-8') as file:
    company_info = json.load(file)

with open('./database/wage_weight.json', 'r', encoding='utf-8') as file:
    wage_weight = json.load(file)

# Select 'test3' user from user_info
selected_user = next(user for user in user_info if user['id'] == 'test3')
user_age = selected_user['user_age']

# Calculate test3_waw for each company in company_info
company_waw_values = {}
for company in company_info:
    company_wage = company['company_wage'] / 10000
    for weight in wage_weight:
        if weight['aged'] == user_age and weight['income1'] <= company_wage < weight['income2']:
            company_waw_values[f"test3_waw_{company['company_name']}"] = weight['weight'] * (
                    selected_user['user_preference_wage'] /
                    (selected_user['user_preference_wage'] + 
                     selected_user['user_preference_commute'] + 
                     selected_user['user_preference_job'])
            )

# Convert to DataFrame for display
df_company_waw = pd.DataFrame(list(company_waw_values.items()), columns=['Company', 'test3_waw'])

df_company_waw
```
