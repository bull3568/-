```python
import json
import pandas as pd
import requests

# Function to get coordinates using Naver Maps API
def get_coordinates(client_id, client_secret, address):
    url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode"
    headers = {
        "X-NCP-APIGW-API-KEY-ID": client_id,
        "X-NCP-APIGW-API-KEY": client_secret
    }
    params = {
        'query': address
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['addresses']:
            x = data['addresses'][0]['x']  # 경도
            y = data['addresses'][0]['y']  # 위도
            return x, y
        else:
            print(f"주소를 찾을 수 없습니다: {address}")
            return None, None
    else:
        print("Failed to retrieve data:", response.status_code, response.text)
        return None, None

# Function to get driving route using Naver Maps API
def get_driving_route(client_id, client_secret, start, goal):
    url = "https://naveropenapi.apigw.ntruss.com/map-direction/v1/driving"
    headers = {
        "X-NCP-APIGW-API-KEY-ID": client_id,
        "X-NCP-APIGW-API-KEY": client_secret
    }
    params = {
        'start': f"{start[0]},{start[1]}",
        'goal': f"{goal[0]},{goal[1]}",
        'option': 'trafast'  # 최적 경로 옵션
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        data = response.json()
        if 'route' in data and 'trafast' in data['route']:
            duration_ms = data['route']['trafast'][0]['summary']['duration']
            duration_minutes = duration_ms / 60000  # Convert ms to minutes
            return duration_minutes
        else:
            print(f"경로 정보를 찾을 수 없습니다: {data}")
            return None
    else:
        print("Failed to retrieve data:", response.status_code, response.text)
        return None

# Load the JSON files
with open('./database/user_info.json', 'r', encoding='utf-8') as file:
    user_info = json.load(file)

with open('./database/company_info.json', 'r', encoding='utf-8') as file:
    company_info = json.load(file)

with open('./database/wage_weight.json', 'r', encoding='utf-8') as file:
    wage_weight = json.load(file)

with open('./database/commute_weight.json', 'r', encoding='utf-8') as file:
    commute_weight = json.load(file)

# Select 'test3' user from user_info
selected_user = next(user for user in user_info if user['id'] == 'test3')
user_age = selected_user['user_age']

# Naver API credentials
client_id = 'gx4wwzfwty'
client_secret = '71tBFr6FCb6qJPGzS5HjLekN8qhPIsBryYehi5d2'

# Get the user's address and coordinates
start_address = selected_user['address']
start_coordinates = get_coordinates(client_id, client_secret, start_address)

# Initialize dictionaries to store route information and commute weights
route_info = {}
commute_weights = {}
test3_cw_values = {}

# Calculate test3_waw for each company in company_info
company_waw_values = []

if start_coordinates:
    for company in company_info:
        goal_address = company['company_address']
        goal_coordinates = get_coordinates(client_id, client_secret, goal_address)
        
        if goal_coordinates:
            travel_time = get_driving_route(client_id, client_secret, start_coordinates, goal_coordinates)
            if travel_time is not None:
                route_key = f'route_info_to_{goal_address}'
                route_info[route_key] = travel_time
                
                # Find commute weight based on travel time
                for cw in commute_weight:
                    if cw['city_name'] == '서울특별시' and cw['time1'] < travel_time <= cw['time2']:
                        commute_weights[route_key] = cw['car_score']
                        
                        # Adjust commute weights based on user preferences
                        test3_cw = cw['car_score'] * (selected_user['user_preference_commute'] /
                                                      (selected_user['user_preference_wage'] + 
                                                       selected_user['user_preference_commute'] + 
                                                       selected_user['user_preference_job']))
                        test3_cw_values[route_key] = test3_cw

                # Calculate test3_waw for each company
                company_wage = company['company_wage'] / 10000
                for weight in wage_weight:
                    if weight['aged'] == user_age and weight['income1'] <= company_wage < weight['income2']:
                        company_waw_values.append({
                            "company_name": company['company_name'],
                            "company_address": goal_address,
                            "test3_waw": weight['weight'] * (
                                selected_user['user_preference_wage'] /
                                (selected_user['user_preference_wage'] + 
                                 selected_user['user_preference_commute'] + 
                                 selected_user['user_preference_job'])
                            ),
                            "route_key": route_key
                        })

# Convert route_info, commute_weights, and test3_cw_values to DataFrames
df_route_info = pd.DataFrame(route_info.items(), columns=['Route', 'Travel Time (minutes)'])
df_commute_weights = pd.DataFrame(commute_weights.items(), columns=['Route', 'Commute Weight'])
df_test3_cw_values = pd.DataFrame(test3_cw_values.items(), columns=['Route', 'test3_cw'])
df_company_waw = pd.DataFrame(company_waw_values)

# Merge the DataFrames
test3_weight = df_company_waw.merge(df_test3_cw_values, left_on='route_key', right_on='Route')

# Select and rename columns to match the desired format
test3_weight = test3_weight[['company_name', 'company_address', 'test3_waw', 'test3_cw']]
test3_weight.rename(columns={'company_name': 'Company', 'company_address': 'Address'}, inplace=True)

# Display the combined DataFrame
test3_weight
```

```python
import json
import pandas as pd

# Load the JSON files
with open('./database/user_info.json', 'r', encoding='utf-8') as file:
    user_info = json.load(file)

with open('./database/company_info.json', 'r', encoding='utf-8') as file:
    company_info = json.load(file)

with open('./database/job_code.json', 'r', encoding='utf-8') as file:
    job_code = json.load(file)

with open('./database/job_weight.json', 'r', encoding='utf-8') as file:
    job_weight = json.load(file)

# Select 'test3' user from user_info
selected_user = next(user for user in user_info if user['id'] == 'test3')
user_previous_job = selected_user['user_previous_job']

# Create a mapping from job codes, assuming job_code.json is a list of dictionaries
# 수정
job_code_mapping = {item['code_5']: item['code'] for item in job_code}
# print(job_code_mapping)
# Create a mapping for job weight
job_weight_mapping = {weight['number_of_matches']: weight['job_match_score'] for weight in job_weight}

# Job matching and calculation for test3_jw
test3_jw_values = []

for company in company_info:
    # 수정
    company_job_code = job_code_mapping.get(company['job_name'], {})
    match_length = 0
    # print(user_previous_job)
    # print(company_job_code)
    for i in range(min(len(user_previous_job), len(company_job_code))):
        if user_previous_job[i] == company_job_code[i]:
            match_length += 1
        else:
            break
    
    job_match_score = job_weight_mapping.get(match_length, 0)
    adjusted_job_weight = job_match_score * (selected_user['user_preference_job'] /
                                             (selected_user['user_preference_wage'] + 
                                              selected_user['user_preference_commute'] + 
                                              selected_user['user_preference_job']))
    test3_jw_values.append({
        "company_name": company['company_name'],
        "test3_jw": adjusted_job_weight
    })

# Convert job weight values to DataFrame
df_test3_jw = pd.DataFrame(test3_jw_values)
```
